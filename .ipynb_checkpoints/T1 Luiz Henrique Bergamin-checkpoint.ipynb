# 2 . Parâmetros: <a name="section_2"></a>

Para o calculo da razão de amortecimento será usado somente uma fração  da vibração medida compreendida do intervalo de tempo ***1.0***(s) ao ***5.0***(s):

```python
atividade1_cut = atividade1.extract((1.0, 5.0), by='time')
```

Onde os dados do intervlado apresentam os seguintes parâmetros:

```python
atividade1_cut.printAttrib()
```

Retornando assim os 3 seguintes graficos de amplitude com o novo intervalo de tempo determinado:

```python
atividade1_cut.plot_time(fig=2);
```

* ##  (a) Razão de amortecimento por decremento logarítmico: <a name="section_2.a"></a>

```python
fs = (atividade1_cut.fs)
Td = (atividade1_cut.Td)


print('Taxa de amostragem (fs):         {0:4.1f}Hz'.format(fs))
print('Tempo de duração da série (Td):  {0:4.1f}s' .format(Td))
```

```python
fn = 1.0                 # natural vibration frequency (1 cycle per second = Hz)
wn = 2*np.pi*fn          # conversion to rad/s
zt = 0.01                # damping ratio of 3% of the critical (try changing this)
u0 = 5.                 # initial displacement of 1 meter
v0 = 0.                  # oscillation starts from rest 

                
t  = np.linspace(0, Td, int(fs*Td))

wD = wn*np.sqrt(1 - zt**2)

e  = np.exp(-zt*wn*t)
uD = e*(u0*np.cos(wD*t) + ((v0 + zt*wn*u0)/wD)*np.sin(wD*t))

plt.figure(2, figsize=(10, 4), clear=True)
plt.plot(t, uD)
plt.plot(t, u0*e, 'r:', t, -u0*e, 'r:')

plt.xlim(  0, Td);  plt.xlabel('time (s)') 
plt.ylim(-10, 10);  plt.ylabel('u(t)') 

plt.grid(True)
```

```python
#Onde uD é um objeto da classe tipo:
print(type(uD), uD.shape)
```

* ## (b) Frequência e razão de amortecimento por ``MRPy.fit_decay()``: <a name="section_2.b"></a>

```python
afit, par = atividade1_cut.fit_decay()
t         = atividade1_cut.t_axis()               # prepare time axis for plotting

plt.figure(3, figsize=(12, 6), clear=True)
plt.plot(t, atividade1_cut[2])
plt.plot(t, afit[2], 'm:')
plt.legend(('original','fitted'))

plt.xlim( 0, 3);  plt.xlabel('time (s)') 
plt.ylim(-3, 3);  plt.ylabel('u(t)') 

plt.grid(True) 

print(par[2,0])

print('Amplitude:     {0:4.2f}m/s2'.format(par[2,0]))
print('Frequency:     {0:4.2f}Hz  '.format(par[2,1]))
print('Damping ratio: {0:4.2f}%   '.format(par[2,2]*100))
print('Phase:         {0:4.2f}rad '.format(par[2,3]))
```
